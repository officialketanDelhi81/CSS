1. You are building a landing page for a small bakery in your city. The page looks fine
on your laptop but when you open it on your phone, the text is too small and the images go
outside the screen. Which CSS technique will you use to fix this and how will you start?
  - To fix this problem, I would use Responsive Web Design (RWD) with CSS Media Queries.

  Steps to start:

  1. Add the viewport meta tag in the HTML:
  2. Use flexible units such as %, vw, vh, rem, and max-width instead of fixed pixel values where possible.
  3. Add a media query for mobile devices:

```python
1. <meta name="viewport" content="width=device-width, initial-scale=1.0">
3. /* Default styles for larger screens */
.container {
    width: 80%;
    margin: auto;
}

img {
    max-width: 100%;
    height: auto;
}

h1 {
    font-size: 2rem;
}

/* Styles for screens 768px or smaller */
@media and screen (max-width: 768px) {
    .container {
        width: 95%;
    }

    h1 {
        font-size: 1.5rem;
    }

    p {
        font-size: 1rem;
    }
}
```

3.  You wrote CSS for a button — first you added a class .btn with blue background,
then added an id #submitBtn with red background to the same button. The button is showing
red instead of blue. Why is this happening?
  - The button appears red because #submitBtn has higher specificity than .btn, so the browser applies the red background from the ID selector instead of the blue background from the class selector.

```python
# html
<button id="submitBtn" class="btn">Submit</button>
# css
.btn {
    background-color: blue;
}

#
(id)submitBtn {
    background-color: red;
}
```

4.  You are building a product card for an online store. You want three cards to sit side
by side on desktop, but stack one below the other on mobile. Which layout system will you use
and how will you make it responsive?
  - To display three product cards side by side on desktop and stack them vertically on mobile, I would use Flexbox with a media query.

       I would use CSS Flexbox and make it responsive with a media query that changes the layout from a horizontal row on desktop to a vertical column on mobile.

```python
# css
.container {
    display: flex;
    gap: 20px;
}

.card {
    flex: 1;
}

/* Mobile view */
@media (max-width: 768px) {
    .container {
        flex-direction: column;
     }
}
```

6. Your senior gives you a design where a 'SALE' label has to sit on the top-right
corner of every product image. The label should overlap the image. Which positioning approach
will you use?
  - I would use absolute positioning for the "SALE" label inside a relatively positioned product container. This allows the label to overlap the image and stay fixed in the top-right corner of each product image.

```python
# html
<div class="product">
    <img src="product.jpg" alt="Product">
    <span class="sale">SALE</span>
</div>
# css
.product {
    position: relative;
    display: inline-block;
}

.sale {
    position: absolute;
    top: 10px;
    right: 10px;
    background: red;
    color: white;
    padding: 5px 10px;
    font-weight: bold;
}
```

8.  What is the difference between Flexbox and CSS Grid? When should you use
each?
  - | **Flexbox**                                                     | **CSS Grid**                                                        |
| --------------------------------------------------------------- | ------------------------------------------------------------------- |
| One-dimensional layout (row **or** column).                     | Two-dimensional layout (rows **and** columns).                      |
| Best for aligning and distributing items in a single direction. | Best for creating complete page layouts with both rows and columns. |
| Easier for small components like menus, buttons, and cards.     | Better for complex layouts like dashboards and web pages.           |
Use Flexbox when you need to arrange items in a single row or a single column.
Example: Navigation bars, card rows, buttons, or form elements.
Use CSS Grid when you need to create a two-dimensional layout with both rows and columns.
Example: Web page layouts, photo galleries, dashboards, and product grids.

```python
#css
'''.container {
    display: flex;
    gap: 20px;
}'''

'''.container {
    display: grid;
    grid-template-columns: repeat(3, 1fr);
    gap: 20px;
}'''
```

It keeps HTML and CSS separate, making the code cleaner.
The same CSS file can be used by multiple web pages.
It is easier to maintain and update.
Browsers can cache the CSS file, improving website performance.
